```python
import matplotlib.pyplot as plt
import numpy as np 
import ast
```

```python
file_DDQN = open("DDQN.txt", "r")
file_DQN = open("DQN.txt", "r")
file_DQN_EP = open("DQN_EP.txt", "r")
file_DQN_EP_SN = open("DQN_EP_SN.txt", "r")
```

```python
DDQN      = file_DDQN.readlines()
DQN       = file_DQN.readlines()
DQN_EP    = file_DQN_EP.readlines()
DQN_EP_SN = file_DQN_EP_SN.readlines()

DDQN_reward = ast.literal_eval(DDQN[0])
DQN_reward = ast.literal_eval(DQN[0])
DQN_EP_reward = ast.literal_eval(DQN_EP[0])
DQN_EP_SN_reward = ast.literal_eval(DQN_EP_SN[0])

DDQN_loss = ast.literal_eval(DDQN[1])
DQN_loss = ast.literal_eval(DQN[1])
DQN_EP_loss = ast.literal_eval(DQN_EP[1])
DQN_EP_SN_loss = ast.literal_eval(DQN_EP_SN[1])
```

```python
length = len(DDQN_reward)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

x = np.linspace(0, length, length)

ax.plot(x, DQN_reward, label='DQN', color='red')
ax.plot(x, DQN_EP_reward, label='DQN with Experience Replay', color='green')
ax.plot(x, DQN_EP_SN_reward, label='DQN with Experience Replay and Separate Network', color='blue')
ax.plot(x, DDQN_reward, label='DDQN', color='purple')
ax.legend()
ax.grid()

ax.set_xlabel("Step from 0 to 25000. Information earned when step % 1000 == 0")
ax.set_ylabel("Last 100 steps average reward.")
ax.set_title("Average reward comparison among DQN and its Friends")

ax.plot()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

x = np.linspace(0, length, length)

ax.plot(x, DQN_loss, label='DQN', color='red')
ax.plot(x, DQN_EP_loss, label='DQN with Experience Replay', color='green')
ax.plot(x, DQN_EP_SN_loss, label='DQN with Experience Replay and Separate Network', color='blue')
ax.plot(x, DDQN_loss, label='DDQN', color='purple')
ax.legend()
ax.grid()

ax.set_xlabel("Step from 0 to 25000. Information earned when step % 1000 == 0")
ax.set_ylabel("Loss earned when step % 1000 == 0")
ax.set_title("Loss Comparison among DQN and its Friends")

ax.plot()
```
